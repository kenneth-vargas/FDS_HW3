```python
def pretty_plot(ax, title, x_label, y_label, title_size=20, label_size=12): 
    ax.set_title(title, size=title_size)
    ax.set_ylabel(y_label, fontsize=label_size)
    ax.set_xlabel(x_label, fontsize=label_size)
```

```python
FIG_SIZE = (8, 8)
sns.set(font_scale=1.1)
```

```python
def is_categorical(datatype): 
    return True if datatype['base'] == 'string' else False

def is_numerical(datatype): 
    return True if datatype['base'] == 'decimal' else False
```

# Data Source

The dataset is a compilation of Form D fillings. The SEC uses this filing to record the offering of securities under Regulation D. Each row of the dataset is either a new filing or an amendment to a previous investment by a private investor to a non-registered security. The data is obtained from the SEC website. This python script downloads and unzips the data. 

```python
# !curl -LO https://www.sec.gov/files/structureddata/data/form-d-data-sets/2022q3_d.zip
# !unzip 2022q3_d.zip
```

```python
# Display data description
f = codecs.open('./2022Q3_d/FormD_metadata.json', 'r', 'utf-8-sig')
data = json.load(f, strict=False)  
# Closing file
f.close()
```

# Data structure

FormD_metadata.json briefly describes each variable. The file also contains information on how the data was stored, like datatype, a required flag, and names used in other documents. 

```python
variable_description=pd.DataFrame.from_dict(data['tables'][2]['tableSchema']['columns'])
variable_description[['name', 'required', 'dc:description']]
```

## Read Offerent.tsv data

The dataset comes in a file with a .tvs extension. In this file type each entry is separated by a tab. Each like in the file represents a new entry. The following scripts loads the data into a pandas Dataframe.  


```python
offering = pd.read_csv('./2022Q3_d/OFFERING.tsv', sep='\t')
offering.head()
```

From the output of offering.dtypes we can see that all numerical values are stored as float64. The SALE_DATE column is a type object and it will be transformed into a DateTime object.

Many of the variables are flags/booleans which will be transformed from a string into a boolean data type. In many cases, the flags which were set to false were just left black as observed by the output of the command offering[cat_variables].isna().sum(). The nan values are filled with False.

## Transform data to boolean type

```python
def transform_to_boolean(column):
    return column.astype('boolean').fillna(False)
```

```python
BOOLEAN_COLUMNS = ['IS40ACT','ISAMENDMENT','YETTOOCCUR', 'MORETHANONEYEAR',
       'ISEQUITYTYPE', 'ISDEBTTYPE', 'ISOPTIONTOACQUIRETYPE',
       'ISSECURITYTOBEACQUIREDTYPE', 'ISPOOLEDINVESTMENTFUNDTYPE',
       'ISTENANTINCOMMONTYPE', 'ISMINERALPROPERTYTYPE', 'ISOTHERTYPE',
       'ISBUSINESSCOMBINATIONTRANS',
       'OVER100RECIPIENTFLAG',
       'SALESCOMM_ISESTIMATE', 
       'FINDERSFEE_ISESTIMATE', 'GROSSPROCEEDSUSED_ISESTIMATE']
```

```python
offering[BOOLEAN_COLUMNS] = offering[BOOLEAN_COLUMNS].apply(lambda x: transform_to_boolean(x), axis=1)
```

# Univariate distribution

```python
def univariate_distribution_analysis(df, column_name, figsize =(8, 8)):
    fig, ax = plt.subplots(nrows=1, figsize=figsize)
    #ax.set_title(column_name)
    sns.violinplot(x=np.log(df[column_name] + 1), ax=ax)
    plt.show()
    
```

```python
for column in NUMERICAL_COLUMNS:
    univariate_distribution_analysis(offering, column)
```

Because of the large variance in the numerical values, we plotted the log of the values. MINIMUMINVESTMENTACCEPTED, TOTALAMOUNTSOLD, TOTALNUMBERALREADYINVESTED, SALESCOMM_DOLLARAMOUNT, GROSSPROCEEDSUSED_DOLLARAMOUNT. Seem to have bimodal distributions. 

FINDERSFEE_DOLLARAMOUNT might have some possible outliers since the violin plot shoes a large max value.

# Missing data

```python
offering.isna().sum()
```

The table above shows the count of missing values in each column. The missing values in many of these columns are because of the type of entry in the table. For example, if an entry is not an amendment, the previous case number should be null. In many of the categorical columns, the value null also represented false, so it was a design decision to set False and an empty value. Many classification columns have a large number of null values which might be because the specific entry did not need any extra clarification. 

Other types of missing data might be a problem. There is no reason why the INVESTMENTFUNDTYPE  should be null unless null means a specific default category unknown to us. SALE_DATE is another important category with empty values. 

Overall most of the important variables seem to not have missing data. However, because of a design decision of the database, it is impossible to know whether some of the null values are the omission of important data, or not applicable.
